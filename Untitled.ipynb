```python
import pandas as pd
import numpy as np
```

```python
training_data = pd.read_csv('training_data.csv', sep = '\t')
training_data = training_data.drop([training_data.columns[0]], axis = 1)
test_data = pd.read_csv('test_data.csv', sep = '\t')
test_data = test_data.drop([test_data.columns[0]], axis = 1)
```

```python
#for plotting purposes
from pandas.tools.plotting import autocorrelation_plot
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt

#fundamental backend
import tensorflow as tf

#keras and scikit-learn as extensions to tensorflow
from keras import initializers
from keras import regularizers
from keras.models import Sequential, load_model
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.recurrent import LSTM
import sklearn
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.metrics import mean_squared_error, confusion_matrix

#for other code optimisation purposes
import math
import datetime
```

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
	n_vars = 1 if type(data) is list else data.shape[1]
	df = pd.DataFrame(data)
	cols, names = list(), list()
	# input sequence (t-n, ... t-1)
	for i in range(n_in, 0, -1):
		cols.append(df.shift(i).drop(['score'], axis = 1))
		names += [('%s(t-%d)' % (training_data.columns.values[j], i)) for j in range(n_vars-1)]
	names += [((training_data.columns.values[j])) for j in range(n_vars)]
	cols.append(df.shift(0))
	# put it all together
	agg = pd.concat(cols, axis=1)
	agg.columns = names
	# drop rows with NaN values
	if dropnan:
		agg.dropna(inplace=True)
	return agg
 
```

```python
# set number of days before prediction to consider, default = 1
lag_seconds = 20
# set number of days to predict in the future
prediction_days = 0

# calculate number of stocks and initialise columns to drop
no_of_params = len(test_data.columns)
drop_col = []

# frame as supervised learning
reframed = series_to_supervised(test_data, lag_seconds + prediction_days-1, 0)

# displays table
display(reframed)
tframed = [] 
for row in reframed.values:
    tframed.append([])
    for param in range(8):
        plist = []
        for days in range(20):
            plist.append(row[param+days*8])
        average = np.mean(plist)
        variance = np.var(plist)
        tframed[-1].append(average)
        tframed[-1].append(variance)
    tframed[-1].append(row[-1])
        
tframed = pd.DataFrame(tframed, columns = ['blink_avg','blink_var','acc_avg', 'acc_var', 'stable_avg', 'stable_var', 'humid_avg', 'humid_var', 'temp_avg', 'temp_var', 'yawn_avg', 'yawn_var', 'disorder_avg', 'disorder_var', 'sec_avg', 'sec_var', 'score'])

# obtain most current data for prediction later
last = tframed.iloc[:,len(drop_col)+1:].values[-prediction_days:]
        

# initialise scaler object
tscaler = MinMaxScaler(feature_range=(0, 1))
scaled = tscaler.fit_transform(tframed.values)

# convert reframed to scaled for easier referencing below
tframed = scaled
```

```python
train_X, train_y = newframed[:, :-1], newframed[:, -1]
test_X, test_y = tframed[:, :-1], tframed[:, -1]
train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
# initialise model
model = Sequential()

# add LSTM layers with set number of neurons, activation and regularizer
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2]),
               activation="tanh",
               kernel_regularizer=regularizers.l2(0.01),
               return_sequences = True))
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2]),
               activation="tanh"))
# add normal layer
model.add(Dense(1))
# set loss function and optimizer
model.compile(loss='mean_squared_error', optimizer='rmsprop')

# fit model with specified epochs and batch size
history = model.fit(train_X, train_y, epochs=500, batch_size=72, validation_data=(test_X, test_y), verbose=2, shuffle=False)
```

```python
# make a prediction
yhat = model.predict(test_X)
test_X = test_X.reshape((test_X.shape[0], test_X.shape[2]))

# invert scaling for predicted values
inv_yhat = np.concatenate((test_X[:, :], yhat), axis=1)
inv_yhat = tscaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,-1]

# invert scaling for actual values
test_y = test_y.reshape((len(test_y), 1))
inv_y = np.concatenate((test_X[:, :], test_y), axis=1)
inv_y = tscaler.inverse_transform(inv_y)
inv_y = inv_y[:,-1]

# calculate RMSE
rmse = math.sqrt(mean_squared_error(inv_y, inv_yhat))
print('Test RMSE: %.3f' % rmse)
```

```python
# extract training and test loss from model history to plot
plt.figure(figsize=(20,15))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.legend(loc="best")
plt.show()
```

```python
# create expected and predicted seqences
expected = []
predicted = []

# code 1, -1 depending on rise/fall
for i in range(0, len(inv_yhat)):
    if inv_y[i] < 5:
        expected.append(1)
    elif inv_y[i] < 8:
        expected.append(2)
    else:
        expected.append(3)
    if inv_yhat[i]< 5:
        predicted.append(1)
    elif inv_yhat[i] <8:
        predicted.append(2)
    else:
        predicted.append(3)

correct = 0
incorrect = 0

for i in range(0,len(inv_yhat)):
    if expected[i] == predicted[i]:
        correct += 1
    else:
        incorrect += 1
correct = correct/len(inv_yhat)
incorrect = incorrect/len(inv_yhat)

# display confusion matrix
matrix = pd.DataFrame()
matrix[""] = [correct, incorrect]
matrix.index = ['correct', 'incorrect']
matrix
```
